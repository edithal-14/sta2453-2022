```python
import numpy as np
from scipy.stats import norm, t
```

```python
n_x = 10000
n_y = 10000
s_x = 5
s_y = 5
```

```python
x_i = np.random.normal(loc=0.2, scale=s_x, size=n_x)
y_i = np.random.normal(loc=0.3, scale=s_y, size=n_y)
```

```python
mu_x = np.mean(x_i)
mu_y = np.mean(y_i)
print((mu_x, mu_y))
```

```python
s = np.sqrt(((n_x - 1)*(s_x ** 2) + (n_y - 1)*(s_y ** 2)) / (n_x + n_y -2))
T = (mu_x - mu_y) / (s * np.sqrt(1/n_x + 1/n_y))
T
```

```python
t.cdf(T, n_x+n_y-2) * 2
```

```python
B = 10000
T_vals = []
for i in range(B):
  x_y = np.random.choice(np.concatenate((x_i, y_i)), size=(n_x + n_y))
  x_i_new = x_y[:n_x]
  y_i_new = x_y[n_x:]
  mu_x = np.mean(x_i_new)
  mu_y = np.mean(y_i_new)
  var_x = np.var(x_i_new)
  var_y = np.var(y_i_new)
  s = np.sqrt(((n_x - 1)*var_x + (n_y - 1)*var_y) / (n_x + n_y - 2))
  t_new = (mu_x - mu_y) / (s * np.sqrt(1/n_x + 1/n_y))
  T_vals.append(t_new)
```

```python
len([i for i in T_vals if np.abs(i) > np.abs(T)]) / B
```

Theoretical: 0.781
