```python
import numpy as np
from scipy.stats import norm, t
```

```python
n_x = 1000
n_y = 2000
```

```python
x_i = np.random.normal(loc=0.2, scale=10, size=n_x)
y_i = np.random.normal(loc=0.3, scale=5, size=n_y)
```

```python
s = np.sqrt(((n_x - 1)*10 + (n_y - 1)*5) / (n_x + n_y -2))
T = (0.2 - 0.3) / (s * np.sqrt(1/n_x + 1/n_y))
T
```

```python
t.cdf(-1, n_x+n_y-2) * 2
```

```python
B = 10000
T_vals = []
for i in range(B):
  x_y = np.random.choice(np.concatenate((x_i, y_i)), size=(n_x + n_y))
  x_i_new = x_y[:n_x]
  y_i_new = x_y[n_x:]
  mu_x = np.mean(x_i_new)
  mu_y = np.mean(y_i_new)
  var_x = np.var(x_i_new)
  var_y = np.var(y_i_new)
  s = np.sqrt(((n_x - 1)*var_x + (n_y - 1)*var_y) / (n_x + n_y - 2))
  t_new = (mu_x - mu_y) / (s * np.sqrt(1/n_x + 1/n_y))
  T_vals.append(t_new)
```

```python
len([i for i in T_vals if np.abs(i) > np.abs(T)]) / B
```
